```python
#import all the things..
#pandas for dataframes
import pandas as pd
#things not yet done?
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
#new libary imports, to create plots and show them
from bokeh.plotting import figure, output_notebook, show 
# color palets
from bokeh.palettes import Spectral4
#So plots can be one above the other
from bokeh.layouts import column
#arrow annotation
from bokeh.models import Arrow, NormalHead, OpenHead, VeeHead
#lable annotation
from bokeh.models.annotations import Label
# title location (graph lables)
from bokeh.models import Title
```

```python
from bokeh.layouts import row
from bokeh.models import CrosshairTool, Span
from bokeh.plotting import figure, show
# rework with new df
years= [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016,
 2017, 2018,]
# creates top plot background
Top = figure(title= 'Yearly Employment Change', title_location='left', width=800, height=250, 
             x_axis_type="linear", tools='hover')

# colors for the viz
color_list = ['deepskyblue', 'mediumblue', 'darkcyan', 'lime', 'forestgreen', 'coral', 'red', 'orange','salmon', 'tomato']

#creates backgoround for second (bottom) plot
botp = figure(title= "Win Rate", title_location= 'left', width = 800, height = 400, tools= 'hover')

#creates words where title_location specifies
#Top.title.text = 'Click on legend entries to mute the corresponding lines'
botp.add_layout(Title(text="Click Citys for Color", align="center"), "below")

#counter for for loop
i =0

#for loop to create plots
for city in  new.groupby('City'):
# creates lines
    Top.line(years, city[1]['Employment% Dif'] , line_width=2, alpha=0.2, muted_alpha=0.9, line_dash=[(i+1), 4],
             legend_label=city[0], line_color = color_list[i])
#creates bars
    botp.vbar(x= years, top= city[1]['Win % Dif Value'], width=0.8, fill_color= color_list[i],
              fill_alpha=0, muted_alpha=0.9, legend_label= city[0], line_width=1.2) 
#adds one to counter each time thru the loop
    i+=1
    
#labels for both charts
data_label = Label(
    x=2008,  # place the label at the same x coordinates as the recession
    y=4,  # place the label at y coordinates 
    x_offset=1,  # offset the label from the circle center by 12 pixels in the x direction
    text="2008", # the recession
    text_baseline="middle",  # align the text vertically to the middle of the circle
    text_color="olive",  # use a standard text property for the text color
)
#puts lable in top chart
Top.add_layout(data_label)
#puts lable in bottom chart
botp.add_layout(data_label)

#creates arrow for both charts
arrow = Arrow(
    end=OpenHead(line_color="black", line_width=3),  # üîÅ try using  OpenHead or VeeHead instead of OpenHead
    line_width=3,
    line_color='black',
    x_start=2008,
    y_start=8,  # the start of the arrow, defined in data coordinates
    x_end=2008,
    y_end=-4,  # the tip of the arrow is at the minimum of the curve
)

#adds arrow to top viz
Top.add_layout(arrow)
#adda arrow to bottom 
botp.add_layout(arrow)

#puts top legend in place
Top.legend.location = "top_left"
# hides lines when legend is clicked
Top.legend.click_policy="mute"

#bottom plot legend
botp.legend.location='top_left'
#hides bars on click (note, this can also be done by changing "muted_alpha")
botp.legend.click_policy='mute'

#crosshairs for top and bottom plots, linked- setup
width = Span(dimension="width", line_dash="dashed", line_width=2)
height = Span(dimension="height", line_dash="dotted", line_width=2)
#add crosshairs
Top.add_tools(CrosshairTool(overlay=[width, height]))
botp.add_tools(CrosshairTool(overlay=[width, height]))

#displays plots, column puts them one on top of other
show(column(Top, botp))
```
