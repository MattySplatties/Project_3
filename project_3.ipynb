```python
MainNFL = Maindf[(Maindf["Sport"] == "NFL")]
MainNBA = Maindf[(Maindf["Sport"] == "NBA")]
MainMLB = Maindf[(Maindf["Sport"] == "MLB")]
MainMLS = Maindf[(Maindf["Sport"] == "MLS")]
MainNHL = Maindf[(Maindf["Sport"] == "NHL")]

FinalMainNFL = pd.merge(MainNFL, Footballdf, on = "Team", how = "left")
FinalMainNBA = pd.merge(MainNBA, BBalldf, on = "Team", how = "left")
FinalMainMLB = pd.merge(MainMLB, MLBdf, on = "Team", how = "left")
FinalMainMLS = pd.merge(MainMLS, MLSdf, on = "Team", how = "left")
FinalMainNHL = pd.merge(MainNHL, Hockeydf, on = "Team", how = "left")


FinalMainMLS = FinalMainMLS.drop(columns = ["Years Played"])
FinalMainMLB = FinalMainMLB.drop(columns = ["Years Played"])
FinalMainNBA = FinalMainNBA.drop(columns = ["Years Played"])
FinalMainNFL = FinalMainNFL.drop(columns = ["Years Played"])
FinalMainNHL = FinalMainNHL.drop(columns = ["Years Played"])

FinalMainNBA["Ties"] = 0
FinalMainMLB["Ties"] = 0

FinalMainMLS = FinalMainMLS.drop(columns = ["LgRank"])
FinalMainNHL = FinalMainNHL.drop(columns = ["Over_Time_Loss"])


FinalMainMLB["Ties"] = FinalMainMLB["Ties"].fillna(0)
FinalMainMLB["Wins"] = FinalMainMLB["Wins"].fillna(0)
FinalMainMLB["Loss"] = FinalMainMLB["Loss"].fillna(0)
FinalMainMLB["Season"] = FinalMainMLB["Season"].fillna(0)
FinalMainMLB["Percentage"] = FinalMainMLB["Percentage"].fillna(0)
FinalMainNHL["Ties"] = FinalMainNHL["Ties"].fillna(0)
FinalMainNHL["Wins"] = FinalMainNHL["Wins"].fillna(0)
FinalMainNHL["Loss"] = FinalMainNHL["Loss"].fillna(0)
FinalMainNHL["Season"] = FinalMainNHL["Season"].fillna(0)
FinalMainNHL["Percentage"] = FinalMainNHL["Percentage"].fillna(0)
FinalMainNBA["Ties"] = FinalMainNBA["Ties"].fillna(0)
FinalMainNBA["Wins"] = FinalMainNBA["Wins"].fillna(0)
FinalMainNBA["Loss"] = FinalMainNBA["Loss"].fillna(0)
FinalMainNBA["Season"] = FinalMainNBA["Season"].fillna(0)
FinalMainNBA["Percentage"] = FinalMainNBA["Percentage"].fillna(0)
FinalMainNFL["Ties"] = FinalMainNFL["Ties"].fillna(0)
FinalMainNFL["Wins"] = FinalMainNFL["Wins"].fillna(0)
FinalMainNFL["Loss"] = FinalMainNFL["Loss"].fillna(0)
FinalMainNFL["Season"] = FinalMainNFL["Season"].fillna(0)
FinalMainNFL["Percentage"] = FinalMainNFL["Percentage"].fillna(0)
FinalMainMLS["Ties"] = FinalMainMLS["Ties"].fillna(0)
FinalMainMLS["Wins"] = FinalMainMLS["Wins"].fillna(0)
FinalMainMLS["Loss"] = FinalMainMLS["Loss"].fillna(0)
FinalMainMLS["Season"] = FinalMainMLS["Season"].fillna(0)
FinalMainMLS["Percentage"] = FinalMainMLS["Percentage"].fillna(0)

FinalMainMLS.columns.values.tolist()

FinalMainMLS = FinalMainMLS[['Sport','City','State','Team','Season','Wins','Loss','Ties','Percentage']]
FinalMainMLB = FinalMainMLB[['Sport','City','State','Team','Season','Wins','Loss','Ties','Percentage']]
FinalMainNFL = FinalMainNFL[['Sport','City','State','Team','Season','Wins','Loss','Ties','Percentage']]
FinalMainNHL = FinalMainNHL[['Sport','City','State','Team','Season','Wins','Loss','Ties','Percentage']]
FinalMainNBA = FinalMainNBA[['Sport','City','State','Team','Season','Wins','Loss','Ties','Percentage']]


FinalMainMLB = FinalMainMLB.astype({'Wins': 'int64'})
FinalMainMLB = FinalMainMLB.astype({'Loss': 'int64'})
FinalMainMLB = FinalMainMLB.astype({'Season': 'int64'})

FinalMainNHL = FinalMainNHL.astype({'Wins': 'int64'})
FinalMainNHL = FinalMainNHL.astype({'Loss': 'int64'})
FinalMainNHL = FinalMainNHL.astype({'Ties': 'int64'})
FinalMainNHL = FinalMainNHL.astype({'Season': 'int64'})

FinalMainNBA = FinalMainNBA.astype({'Wins': 'int64'})
FinalMainNBA = FinalMainNBA.astype({'Loss': 'int64'})
FinalMainNBA = FinalMainNBA.astype({'Season': 'int64'})

df1 = [FinalMainNHL, FinalMainNFL, FinalMainMLB, FinalMainMLS, FinalMainNBA]
FinalMain = pd.concat(df1)

FinalMain

```
