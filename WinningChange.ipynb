```python
#import all the things..
#pandas for dataframes
import pandas as pd
#things not yet done?
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
#new libary imports, to create plots and show them
from bokeh.plotting import figure, output_notebook, show 
# color palets
from bokeh.palettes import Spectral4
#So plots can be one above the other
from bokeh.layouts import column
#arrow annotation
from bokeh.models import Arrow, NormalHead, OpenHead, VeeHead
#lable annotation
from bokeh.models.annotations import Label
# title location (graph lables)
from bokeh.models import Title, Span, CrosshairTool
# tool merging
from bokeh.layouts import gridplot

from bokeh.transform import factor_cmap
```

```python
#csv from db, now with year change values
new= pd.read_csv('../Clean_Data/EmploymentWinsDif.csv')
Prod =pd.read_csv('../Clean_Data/ProductivityWinsDif.csv')
gdp = pd.read_csv('../Cean_Data/PerCapYearWinsDif.csv')
```

```python
# rework with new df
years= [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016,
 2017, 2018,]

# colors for the viz
color_list = ['deepskyblue', 'mediumblue', 'darkcyan', 'lime', 'forestgreen', 'coral', 'red', 'orange','salmon', 'tomato']

#interactive tools for plot
TOOLS = "hover, pan, help, box_zoom, reset"

# creates employment plot background
empl = figure(title= 'Yearly Employment Change', title_location='left', width=400, height=300, 
             x_axis_type="linear", tools= TOOLS)

#create GDP background
GDP =  figure(title= 'Yearly GDP Change', title_location='left', width=400, height=300, 
             x_axis_type="linear", tools= TOOLS)


#creates backgoround for win plot
win = figure(title= "Win Rate", title_location= 'left', width = 400, height = 300, tools= TOOLS)

#create background for productivity
prod = figure(title= "Labor Productivity", title_location= 'left', width = 400, height = 300, tools= TOOLS)

#creates words where title_location specifies- in this case below last plot
GDP.add_layout(Title(text="Click Citys for Color", align="center"), "below")

#counter for for loop
i =0

#for loop to create plots
for city in  new.groupby('City'):
# creates lines for employment
    empl.line(years, city[1]['Employment% Dif'] , line_width=2, alpha=0, muted_alpha=0.9, line_dash=[(i+1), 4],
             legend_label=city[0], line_color = color_list[i])
#creates bars for wins
    win.vbar(x= years, top= city[1]['Win % Dif Value'], width=0.8, fill_color= color_list[i],
              fill_alpha=0, muted_alpha=0.9, legend_label= city[0], line_width=1.2) 
#adds one to counter each time thru the loop
    i+=1
# 
# creates productivity lines
i=0
for city in  Prod.groupby(by= 'City'):
    prod.line(years, city[1]['Productivity% Dif Value'] , line_width=2, alpha=0, muted_alpha=0.9, line_dash=[(i+1), 4],
             legend_label=city[0], line_color = color_list[i])
    i+=1
    
    
# creates gdp lines
i=0
for city in  gdp.groupby('City'):
    GDP.line(years, city[1]['GDP_per_capita% Dif'] , line_width=2, alpha=0, muted_alpha=0.9, line_dash=[(i+1), 4],
             legend_label=city[0], line_color = color_list[i])
    i+=1

#labels for charts
data_label = Label(
    x=2008,  # place the label at the same x coordinates as the recession
    y=4,  # place the label at y coordinates 
    x_offset=1,  # offset the label from the circle center by 12 pixels in the x direction
    text="2008", # the recession
    text_baseline="middle",  # align the text vertically to the middle of the circle
    text_color="olive",  # use a standard text property for the text color
)
#puts lable in Employment chart
empl.add_layout(data_label)
#puts lable in winnig rate chart
win.add_layout(data_label)
#label for productivity
prod.add_layout(data_label)
#add lable for gdp
GDP.add_layout(data_label)

#creates arrow for both charts
arrow = Arrow(
    end=OpenHead(line_color="black", line_width=3),  # üîÅ try using  OpenHead or VeeHead instead of OpenHead
    line_width=3,
    line_color='black',
    x_start=2008,
    y_start=8,  # the start of the arrow, defined in data coordinates
    x_end=2008,
    y_end=-4,  # the tip of the arrow is at the minimum of the curve
)

#adds arrow to employment viz
empl.add_layout(arrow)
#adda arrow to winning
win.add_layout(arrow)
#add arrow to Productivity
prod.add_layout(arrow)
#add arrow to gdp
GDP.add_layout(arrow)


#puts top legend in place (employment)
empl.legend.location = "top_left"
# hides lines when legend is clicked (employment) 
#(note: since transparancy is set to 0 for start and 0.9 "on mute" this createst the 'show on mute' effect )
empl.legend.click_policy="mute"


#plot legend (winning)
win.legend.location='top_left'
#hides bars on click (note, since transparancy is set to 0 for start and 0.9 "on mute" this createst the 'show on mute' effect
win.legend.click_policy='mute'

#plot legend (productivity)
prod.legend.location='top_left'
#hides line on click (note, since transparancy is set to 0 for start and 0.9 "on mute" this createst the 'show on mute' effect
prod.legend.click_policy='mute'

#plot legend (productivity)
GDP.legend.location='top_left'
#since transparancy is set to 0 for start and 0.9 "on mute" this createst the 'show on mute' effect
GDP.legend.click_policy='mute'


#crosshairs for plots, linked- setup
width = Span(dimension="width", line_dash="dashed", line_width=2)
height = Span(dimension="height", line_dash="dotted", line_width=2)
#add crosshairs
empl.add_tools(CrosshairTool(overlay=[width, height]))
win.add_tools(CrosshairTool(overlay=[width, height]))
prod.add_tools(CrosshairTool(overlay=[width, height]))
GDP.add_tools(CrosshairTool(overlay=[width, height]))

# make a grid
grid = gridplot([empl, win, prod,GDP], ncols=2)

#displays plots, column puts them one on top of other
show(column(grid))
```
